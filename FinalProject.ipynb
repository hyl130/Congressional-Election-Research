## Background & Prior Work

Voters, as the saying goes, don’t bite the hand that feeds them. Common wisdom and several studies show that the economy is a major factor in how people vote, and that if the economy is doing well, voters reward the incumbent party [^one]. Political journalism often speaks of economic metrics such as economic growth percentages and unemployment rates as litmus indicators of a government’s future political success [^two]. Voter behavior prediction is at the core of all politics, and as such there have been countless studies on this topic. There have been entire books written on the relationship between the economy and voter behavior with analysis into factors such as inflation, unemployment, and income, studies in what triggers voting based on the economy, and queries into whether voter’s judgments are based on past economic performance or future policy promises [^three]. Results from one particular study conducted by Lewis-Beck and Stegmaier compare the impact of economics on elections in US and other European countries, displaying how although voting behavior does not rely exlusively on economic performance, voters generally weigh economic issues more heavily than any others regardless of the democracy they vote in [^four].

Indeed there have been research on the impacts of the economy on presidential elections throughout history which echo the goals of our research. A study conducted by Guntermann et al investigate whether voters have held incumbent party accountable for economic circumstances throughout US history, regardless of the incumbents' impact on the economy. Their results found evidence for economic voting, in which voters reward or punish incumbents based on economic growth preceding the election, even in times of the early United States where American economic and electoral systems functioned differently, and that the economy also influences the extent of the victories in vote share [^five]. They discovered that incumbent presidents are much more likely to run and win in an advantageous economic context than in a declining one. With much prior research being done on economic influences in US presidential elections, we attempt to see if the commonly upheld logic stays valid in regards to congressional elections, as well as seeing if the magnitude of the economic prosperity or turmoil correlates with the magnitude of political consolidation or turnover. Our project aims to explore if economic conditions influence whether voters vote for or against the incumbent president's party members in Congress during midterm elections.

## Crude Oil Cleaning

The original crude oil data frame is in a suitable format for analysis, so we leave it as is. We adjust the data frame to only include monthly dates from January 1950 to March 2022. This time range will be used for all of the variables being analyzed so that we may compare economic and political variables across each other within the same time frames. Since the volatility of crude oil prices is presumed to be greatly influenced by inflation in the past decades, we acquire the consumer price index data and divide the nominal crude oil price in a specified month by the ratio of the consumer price index in that month to the consumer price index in a base period to remove the effects of inflation. By adjusting for inflation, we expect the resulting real oil price to give a more accurate sense of what is happening to the price of the commodity. This, in turn, allows for more meaningful comparisons of oil prices over time without inflation also being a major factor in the price change. 

```python
# create a dataframe to include the datas from January 1950 to March 2022
crude_oil_Y = crude_oil.iloc[48:]
crude_oil_Y = crude_oil_Y.rename(columns={"WTISPLC": "Crude Oil"}).reset_index(drop=True)
crude_oil_m = crude_oil.iloc[24:]
crude_oil_m = crude_oil_m.rename(columns={"WTISPLC": "Crude Oil"}).reset_index(drop=True)

cpi_m = cpi.iloc[12:-1]
cpi_m = cpi_m.rename(columns={'CPIAUCSL':'CPI'}).reset_index(drop=True)

oil_price = crude_oil_m['Crude Oil']
cpi_val = cpi_m.CPI
base_cpi = cpi_val[len(cpi_m)-1]

# adjust the crude oil prices according to the cpi values
adjusted_oil = []
for i in range(len(crude_oil_m)):
    adjusted_oil.append(oil_price[i]*base_cpi/cpi_val[i])
    
crude_oil_m['adjusted oil'] = adjusted_oil
Date = crude_oil_Y['DATE']

crude_oil_m
```

### GDP Growth Rate Cleaning

To clean the GDP dataset, we organize the data frame to display values from January 1950 to March 2022. The GDP dataset only represents absolute values from which does not depict an accurate understanding of the economic environment and wellbeing. Therefore, we use the dataset to calculate the real GDP growth rates, which removes the impact of inflation, in order to better reflect the general health of the economy in our research and add them to a new column next to the original GDP values in the data frame. We obtain the formula for computing the real GDP growth rates  from Bureau of Economic Analysis (https://www.bea.gov/help/faq/463). Using this formula ensures that the growth rate of real GDP data is a more accurate indicator of the actual growth of production and the output of the economy that is no longer distorted effects by changes in prices. Furthermore, since this dataset also does not have monthly data, the rows of yearly real GDP growth rates are repeated for each of the 12 months so that it is in a usable formate for combining with other datasets and observing any possible relationships between the variables. 

```python
# create a dataframe to include the dates from January 1950 to March 2022
gdp = gdp.rename(columns={"GDPC1": "GDP"})

gdp_q = gdp.iloc[11:]
gdp_two_year = gdp.iloc[4:]

GDP_growth_Q = []
GDP_growth_TY = []
total_quarters = len(gdp_q['GDP'])
GDP_Q = gdp_q['GDP'].tolist()
GDP_TY = gdp_two_year['GDP'].tolist()

for i in range(0,total_quarters-1):
    
    # Get the annulized ratio in GDP between each consecutive months 
    annual_r = (GDP_Q[i+1]/GDP_Q[i])**4
    
    # Get the growth rate of real GDP
    growth_r = (annual_r-1)*100
    GDP_growth_Q.append(growth_r)
    
gdp_q = gdp_q.iloc[1:,:]
gdp_q['GDP growth'] = GDP_growth_Q
gdp_q = gdp_q.reset_index(drop = True)

rep_gdp = pd.DataFrame(np.repeat(gdp_q.values, 3, axis=0), columns=gdp_q.columns)
rep_gdp['DATE'] = Date

rep_gdp
```

In addition, since the election date will always be in November, we will split our data on October in every election year for our calculation.

```python
# Annualized GDP rate
total_terms = len(gdp_two_year['GDP'])
for j in range(0, int(total_terms/8)):
    average_r = (GDP_TY[8*(j+1)]/GDP_TY[8*j])**(0.5)
    average_g = (average_r-1)*100
    GDP_growth_TY.append(average_g)

# Inflation rate
average_inflation_rate = []
for i in range( 0, len(inflation_rate)-1, 2):
    rate = ((1+ inflation_rate['inflation rate'].iloc[i]) * (1+inflation_rate['inflation rate'].iloc[i+1]) ) - 1
    average_inflation_rate.append( rate * 100 )

# S&P 500 Change
snp500_m = result_merged['S&P500 Total Value']
snp500_difference = [22.73]
for i in range(int(len(snp500_m)/24)-1):
    rate = (snp500_m[24*(i+1)+9] - snp500_m[24*i+9])/snp500_m[24*i+9]
    snp500_difference.append( rate*100 )

# Unemployment Rate Change
unemployment_change = [0.5]
for i in range(int(len(unemployment)/24)-1):
    uemployment_rate = unemployment['Unemployment Rate'][24*(i+1)+9]-unemployment['Unemployment Rate'][24*i+9]
    unemployment_change.append(uemployment_rate)

# Inflation adjusted Crude Oil Price Change
oil_change = []
period = int(len(crude_oil_m)/24-1)
oil_price_m = crude_oil_m['adjusted oil']
for i in range(period):
    oil_change.append((oil_price_m[24*(i+1)+9]-oil_price_m[24*i])/oil_price_m[24*i+9]*100)
```

## Political Party Division Cleaning

To clean the government party division dataset, we remove all the unnecessary columns and keep the data for Senate and House for both Democratic and Republican parties. Using this data will allow us to obtain the number of seats in each of the major two parties. 
Since the we are focusing on the party divisions in Congress based on the initial midterm election outcomes, we are unable to account for the fact the actual number of senators representing a particular party often changes during a Congress term as well due to reasons such as death, resignation, or members switching parties as well. Since Independents are not affiliated with a political party, we are not including them as part of the dataset and are only including the number of seats for the Democratic and Republican parties at the start of each Congress in the data frame. Other irrelevant numbers are removed and the values are then converted to integers to ensure the Data columns are standardized. The number of seats for the corresponding Democratic or Republican party in the Senate and the House is summed to find the number of seats per party in Congress. Using these values, we calculate the change in seats between the Congress terms for each party. The reason for doing this is to observe the number of seats that are won or lost by each party in elections. From there, we can examine whether there is a relationship between the economic trends and the performance of the President's party during elections.

```python
# remove unnecessary columns and reformat the years
party_seats = party_seats.drop(columns=['Congress'])
party_seats[['Start','End']] = party_seats['Years'].str.split('–', expand=True)
party_seats = party_seats.drop(columns='Years')
party_seats.columns = ['Senate_Total','Senate_Dem','Senate_Rep','Senate_Others','Senate_Vacancies','House_Total','House_Dem','House_Rep','House_Others','House_Vacancies','President','Start','End']

# Drop columns for independent party seats, seat vacancies, and President
party_seats = party_seats.drop(columns=['Senate_Others', 'Senate_Vacancies', 'House_Others', 'House_Vacancies', 'President'])

# Replace seat changes with the number of original seats at the start of the Congress
party_seats['Senate_Dem'] = party_seats['Senate_Dem'].replace('50[28]', 50)
party_seats['Senate_Dem'] = party_seats['Senate_Dem'].replace('56–58[33]', 56)
party_seats['Senate_Dem'] = party_seats['Senate_Dem'].replace('46/47', 46)
party_seats['Senate_Dem'] = party_seats['Senate_Dem'].replace('45/46', 45)
party_seats['Senate_Dem'] = party_seats['Senate_Dem'].replace('46–48[37]', 46)
party_seats['Senate_Rep'] = party_seats['Senate_Rep'].replace('50/49[29]', 50)
party_seats['Senate_Rep'] = party_seats['Senate_Rep'].replace('40–42[34]', 40)
party_seats['Senate_Rep'] = party_seats['Senate_Rep'].replace('50-52', 50)
party_seats['Senate_Rep'] = party_seats['Senate_Rep'].replace('53/52', 53)
party_seats['Senate_Rep'] = party_seats['Senate_Rep'].replace('51/50[38]', 51)

# Add up the total seats for each party in the Congress
party_seats['Senate_Dem'] = party_seats['Senate_Dem'].astype(int)
party_seats['Senate_Rep'] = party_seats['Senate_Rep'].astype(int)

party_seats['Dem_Total'] = party_seats['Senate_Dem'] + party_seats['House_Dem']
party_seats['Rep_Total'] = party_seats['Senate_Rep'] + party_seats['House_Rep']

party_seats = party_seats.drop(columns=['Senate_Total', 'House_Total'])
 
# Calculate the difference in seats between terms for both parties in the Senate and House
party_seats['Dem_change'] = party_seats['Dem_Total'].diff()
party_seats['Rep_change'] = party_seats['Rep_Total'].diff()

# Drop first row which has no change value
party_seats = party_seats.iloc[2:,:].reset_index(drop=True)

party_seats.head()
```
